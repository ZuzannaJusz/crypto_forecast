Bitcoin

```python
# import dataset
bitcoin_df = pd.read_csv(r"C:\Users\Zuzanna\Desktop\Nowy_folder\projects\crypto_forecast\archive\Bitcoin\BTC-USD.csv")
bitcoin_df.head()
                         
```

```python
# get info about NA values
bitcoin_df.info()
bitcoin_df.isna().sum()
```

```python
# filling missing data
bitcoin_df.fillna(method='ffill', inplace=True)
```

```python
# checking NA values again
bitcoin_df.isna().sum()
```

```python
# checking for duplicates
bitcoin_df.duplicated().sum()
```

```python
# rename columns
bitcoin_df.rename(columns = 
    {"Date":"date",
    "Open":"open",
    "High":"high",
    "Low":"low",
    "Close":"close",
    "Adj Close":"adj_close",
    "Volume":"volume"
}, inplace=True
                 )
bitcoin_df.head()
```

```python
# change data format
bitcoin_df['date'] = pd.to_datetime(bitcoin_df.date)
```

```python
last6months_bitcoin = bitcoin_df[bitcoin_df['date'] > '03-2021']
last6months_cardano = cardano_df[cardano_df['date'] > '03-2021']
last6months_dogecoin = dogecoin_df[dogecoin_df['date'] > '03-2021']
last6months_ethereum = ethereum_df[ethereum_df['date'] > '03-2021']
```

```python
fig = plt.figure(figsize = (7,7))
fig.suptitle("Porwnanie close i open z okresu od marca 2021 do wrzenia 2021", fontsize=16)
fig.tight_layout()

plt.subplot(4, 1, 1)
plt.plot(last6months_bitcoin['date'], last6months_bitcoin['close'])
plt.plot(last6months_bitcoin['date'], last6months_bitcoin['open'])
plt.legend(["C", "O"])
plt.title("Bitcoin")

plt.subplot(4, 1, 4)
plt.plot(last6months_ethereum['date'], last6months_ethereum['close'], color="blue")
plt.plot(last6months_ethereum['date'], last6months_ethereum['open'], color="yellow")
plt.legend(["C", "O"])
plt.title("Ethereum")

plt.subplot(4, 1, 2)
plt.plot(last6months_cardano['date'], last6months_cardano['close'], color="black")
plt.plot(last6months_cardano['date'], last6months_cardano['open'], color="grey")
plt.legend(["C", "O"])
plt.title("Cardano")

plt.subplot(4, 1, 3)
plt.plot(last6months_dogecoin['date'], last6months_dogecoin['close'], color="orange")
plt.plot(last6months_dogecoin['date'], last6months_dogecoin['open'], color="green")
plt.legend(["C", "O"])
plt.title("Dogecoin")



plt.show()
```
